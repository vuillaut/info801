```python
from IPython.core.magic import register_line_magic

@register_line_magic
def shorterr(line):
    """Show only the exception message if one is raised."""
    try:
        output = eval(line)
    except Exception as e:
        print("\x1b[31m\x1b[1m{e.__class__.__name__}: {e}\x1b[0m".format(e=e))
    else:
        return output
    
del shorterr
```

## Exercise 3

Plot $f(x) = sin(x)$ and $g(x) = cos(x)$ for $x\in[-10, 10]$ in the same plot (overlay).

```python
x = np.linspace(-10, 10, num=200)
f = np.sin(x)
g = np.cos(x)
plt.plot(x, f, label='sin', linestyle='--', color='red', linewidth=2, alpha=0.5, marker='o')
plt.plot(x, g, label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Exercise 6

Plot the functions $f_i(x) = x^{i + 1}$ for $i \in [0, 10]$ in different colours. Show the function definition in the y-label!

```python
x = np.linspace(-1, 1)
for i in range(10):
#     print(i)
    y = x**(1+i)
    plt.plot(x, y, label="power = {}".format(1+i)) # label = f"power = {i+1}"
plt.legend(bbox_to_anchor=(1.1, 1));
```

## Exercise 8

Create a scatterplot from some random data (try `np.random.multivariate_normal`)!

```python
x, y = np.random.multivariate_normal(np.zeros(1000), np.ones((1000, 1000)), size=(2))
# x = np.random.normal(1000)
# y = np.random.normal(1000)
plt.scatter(x, y, alpha=0.4)
plt.axis('equal')
```

## Exercise 9

Create a histogram of a normal distribution.

```python
x = np.random.normal(loc=10, scale=1, size=100000)
y = np.random.normal(loc=10, scale=1, size=100000)
plt.hist(x, bins=100);
plt.show()

plt.hist2d(x, y, bins=100);
plt.colorbar()
plt.axis('equal')
```
